```python
#Simon Tice AI Assignment 1
class WaterJugPuzzle():
  def __init__(self):

    #capacities
    self.CAP_FIVE = 5
    self.CAP_THREE = 3

    # initial empty state
    self.five = 0
    self.three = 0

    #backups for saving and undoing
    self.storeFive = None
    self.storeThree = None


  #state methods----

  #save previous state
  def saveState(self):
    self.storeFive = self.five
    self.storeThree = self.three

  #undos current state to previous state
  def undoState(self):
    self.five = self.storeFive
    self.three = self.storeThree

  #is the state valid
  def stateOk(self):
    return(
        self.five >= 0 and
        self.five <= self.CAP_FIVE and
        self.three >= 0 and
        self.three <= self.CAP_THREE
    )

  #prints the current states
  def printState(self):
    print("===> 5L: {five}, 3L: {three}".format(five=self.five, three=self.three))


  #rules----

  #fill five L
  def fillFive(self):
    self.saveState()
    if self.five == self.CAP_FIVE:
        return False #cant fill if already full
    self.five = self.CAP_FIVE #fill
    if not self.stateOk(): #if the state is not valid undo
        self.undoState()
        return False
    return True


  #fill three L
  def fillThree(self):
    self.saveState()
    if self.three == self.CAP_THREE:
        return False #cant fill if already full
    self.three = self.CAP_THREE #fill
    if not self.stateOk(): #if the state is not valid undo to saved
        self.undoState()
        return False
    return True

  #empty five L
  def emptyFive(self):
      self.saveState()
      if self.five == 0:
          return False #cant empty if its already empty
      self.five = 0 #empty
      if not self.stateOk(): #if the state created is invalid undo
        self.undoState()
        return False
      return True

  #empty three L
  def emptyThree(self):
      self.saveState()
      if self.three == 0:
          return False #cant empty if its already empty
      self.three = 0 #empty
      if not self.stateOk(): #if the state created is invalid undo
          self.undoState()
          return False
      return True

  #pouring water from 3L to 5L
  def threeToFive(self):
    self.saveState()
    if self.three == 0 or self.five == self.CAP_FIVE:
          return False #if 3 is empty or 5 is full cant pour
    freeSpace = self.CAP_FIVE - self.five #how much free space is in 5
    if self.three <= freeSpace:
        amount = self.three #there is enough space to pour all of 3 to 5

    else:
        amount = freeSpace #not enough space for all of 3 so just fill 5

    self.five += amount #add to 5
    self.three -= amount #take from 3
    if not self.stateOk(): #undo if invalid state
        self.undoState()
        return False
    return True

  #pouring water from 5L to 3L
  def fiveToThree(self):
    self.saveState()
    if self.five == 0 or self.three == self.CAP_THREE:
          return False #if 5 is empty or 3 is full cant pour
    freeSpace = self.CAP_THREE - self.three #how much free space is in 3
    if self.five <= freeSpace:
        amount = self.five #there is enough space to pour all of 5 to 3

    else:
        amount = freeSpace #not enough space for all of 5 so just fill 3

    self.five -= amount #take from 5
    self.three += amount #add to 3
    if not self.stateOk(): #undo if invalid state
        self.undoState()
        return False
    return True


w = WaterJugPuzzle()
w.printState() #start state

print("-----------------------")
w.fillFive() #fill jug 5 fully
w.printState()

print("-----------------------")
w.fiveToThree() #pour 3L from jug 5 to jug 3
w.printState()

print("-----------------------")
w.emptyThree() #empty jug 3 out
w.printState()

print("-----------------------")
w.fiveToThree() #pour the 2L from jug 5 to jug 3
w.printState()

print("-----------------------")
w.fillFive() #fully fill jug 5
w.printState()

print("-----------------------")
w.fiveToThree() #pour 1L from jug 5 to jug 3
w.printState()

```
