```python
#HW2 problem 2 Simon Tice
from collections import deque

#undirected graph to show relations
graph = {
  "Adam" : ["Maria"],
  "Maria" : ["Adam", "Maya", "Sophia", "David"],
  "David" : ["Maria", "Maya"],
  "Maya" : ["Maria", "Sophia", "David"],
  "Sophia" : ["Maria", "Maya"],
}

def PotentialFriends(graph, user):
  visited = set() #keep track of visited nodes
  queue = deque([(user, 0)])  # stores node and level in tuple
  visited.add(user)

  while queue: #explore the queue while its not empty
    vertex, level = queue.popleft() #dequeue node and its level

    if level > 1: #if the level is > 1 it is not a direct friend so its a possible friend so they are all printed
      print(vertex, end=" ")

    if level <= 1: #if the level is <= its either the start or a direct friend so we keep searching
      for neighbor in graph[vertex]:
          if neighbor not in visited:
              #enqueue child nodes and mark as visited
              visited.add(neighbor)
              queue.append((neighbor, level+1)) #add one to level to keep track if its a direct friend


PotentialFriends(graph, 'Sophia')
print("\n")
PotentialFriends(graph, 'David')
print("\n")
PotentialFriends(graph, 'Adam')
```
